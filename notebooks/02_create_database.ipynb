```python
import os
import pandas as pd
import sqlite3

os.makedirs("../database", exist_ok=True)

df = pd.read_csv("../data/processed_job_posts.csv")

conn = sqlite3.connect("../database/job_market.db")
df.to_sql("job_posts", conn, if_exists="replace", index=False)
conn.close()

print("Database created successfully.")
```
