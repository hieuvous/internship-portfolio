```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/processed_job_posts.csv")

print("Shape:", df.shape)
display(df.head())

# 1. Role distribution
role_counts = df["role_group"].value_counts()
display(role_counts)

plt.figure(figsize=(10, 5))
role_counts.sort_values().plot(kind="barh")
plt.title("Number of Job Posts by Role Group")
plt.xlabel("Number of Job Posts")
plt.ylabel("Role Group")
plt.show()

# 2. Top skills
skill_cols = [col for col in df.columns if col.startswith("skill_")]

skill_counts = df[skill_cols].sum().sort_values(ascending=False)
skill_counts_clean = skill_counts.copy()
skill_counts_clean.index = (
    skill_counts_clean.index
    .str.replace("skill_", "", regex=False)
    .str.replace("_", " ", regex=False)
)

display(skill_counts_clean)

top_10_skills = skill_counts_clean.head(10)

plt.figure(figsize=(10, 5))
top_10_skills.sort_values().plot(kind="barh")
plt.title("Top 10 Required Skills")
plt.xlabel("Number of Job Posts")
plt.ylabel("Skill")
plt.show()

# 3. Skill demand by role
role_skill_pct = df.groupby("role_group")[skill_cols].mean() * 100

role_skill_pct_clean = role_skill_pct.copy()
role_skill_pct_clean.columns = (
    role_skill_pct_clean.columns
    .str.replace("skill_", "", regex=False)
    .str.replace("_", " ", regex=False)
)

main_skills = [
    "sql", "python", "excel", "power bi",
    "pandas", "machine learning", "etl", "pytorch"
]

available_main_skills = [skill for skill in main_skills if skill in role_skill_pct_clean.columns]
role_skill_main = role_skill_pct_clean[available_main_skills]

display(role_skill_main)

plt.figure(figsize=(12, 6))
plt.imshow(role_skill_main, aspect="auto")

plt.xticks(range(len(role_skill_main.columns)), role_skill_main.columns, rotation=45, ha="right")
plt.yticks(range(len(role_skill_main.index)), role_skill_main.index)

plt.colorbar(label="% of job posts mentioning skill")
plt.title("Skill Demand by Role Group")

for i in range(role_skill_main.shape[0]):
    for j in range(role_skill_main.shape[1]):
        value = role_skill_main.iloc[i, j]
        plt.text(j, i, f"{value:.0f}", ha="center", va="center")

plt.tight_layout()
plt.show()

# 4. Top companies
top_companies = df["company"].value_counts().head(10)

plt.figure(figsize=(10, 5))
top_companies.sort_values().plot(kind="barh")
plt.title("Top 10 Companies by Number of Job Posts")
plt.xlabel("Number of Job Posts")
plt.ylabel("Company")
plt.show()
```
